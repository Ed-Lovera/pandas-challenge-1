## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

---
### Create a column that calculates the subtotal for each line using the unit_price and the qty. (6 points)

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df["subtotal"] = df["unit_price"] * df["qty"]

print("Preview of the new 'subtotal' column:")
df[["unit_price", "qty", "subtotal"]]
```

---
### Create a column for shipping price. Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under. (6 points)

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
standard_ship_rate = 10.00
discount_ship_rate = 7.00

# Create a column for the total_weight of each order.
df["total_weight"] = df["unit_weight"] * df["qty"]

# Set the shipping rate for all orders.
df["shipping_rate"] = standard_ship_rate

# Set the shipping rate for orders over 50 pounds.
df.loc[df["total_weight"] > 50, "shipping_rate"] = discount_ship_rate

# Calculate the shipping price for each line.
df["shipping_price"] = df["shipping_rate"] * df["total_weight"]

print("Preview of the new columns and their input columns:")
df[["unit_weight", "qty", "total_weight", "shipping_rate", "shipping_price"]]
```

---
### Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%. (6 points)

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df["total_price"] = round((df["subtotal"] + df["shipping_price"]) * 1.0925, 2)

print("Preview of the new 'total_price' column and it's input columns:")
df[["subtotal", "shipping_price", "total_price"]]
```

---
### Create a column for the cost of each line using unit cost, qty, and shipping price (assume the shipping cost is exactly what is charged to the client). (6 points)

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
df["order_cost"] = df["unit_cost"] * df["qty"] + df["shipping_price"]

print("Preview of the new 'order_cost' column and it's input columns:")
df[["unit_cost", "qty", "shipping_price", "order_cost"]]
```

---
### Create a column for the profit of each line using line cost and line price. (6 points)

```python
# Create a column for the profit of each line using line cost and line price
df["order_profit"] = df["total_price"] - df["order_cost"]

print("Preview of the new 'order_profit' column and it's input columns:")
df[["total_price", "order_cost", "order_profit"]]
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
for client in most_client_ids:
    client_df = df.loc[df["client_id"] == client]
    client_total = client_df["total_price"].sum()
    print(f"Client {client} total: {client_total:.2f}")
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit.
summary_df = pd.DataFrame({
    "total_units": df.groupby("client_id")["qty"].sum(),
    "total_shipping": df.groupby("client_id")["shipping_price"].sum(),
    "total_revenue": df.groupby("client_id")["total_price"].sum(),
    "total_cost": df.groupby("client_id")["line_cost"].sum(),
    "total_profit": df.groupby("client_id")["line_profit"].sum()
})
summary_df = summary_df.loc[most_client_ids]
summary_df.reset_index(inplace=True)
summary_df
```

```python
# Format the data and rename the columns to names suitable for presentation.
summary_df.rename(lambda x: x.title().replace("_", " "), axis=1, inplace=True)
summary_df.rename(columns={"Client Id": "Client ID"}, inplace=True)
summary_df.rename(columns={"Total Units": "Units"}, inplace=True)

# Define the money columns.
money_columns = [col for col in summary_df.columns if 'Total' in col]

# Define a function that converts a dollar amount to millions.
def currency_format_millions(value):
    return f"{value / 1000000:.6f}"

# Apply the currency_format_millions function to only the money columns. 
for col in money_columns:
    summary_df[col] = summary_df[col].apply(currency_format_millions)

# Rename the columns to reflect the change in the money format. 
summary_df.rename(columns={col: f"{col} (millions)" for col in money_columns}, inplace=True)
summary_df

```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
summary_df_sorted = summary_df.sort_values("Total Profit (millions)", ascending=False)
summary_df_sorted
```
